```python
a = 10
b = 20
sum_result = a + b
print("Sum:", sum_result)
```

## Problem 3: Subtraction, Multiplication, and Division of Two Numbers

```python
a = 20
b = 4
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
```

## Problem 4: Check if a Number is Even or Odd

```python
num = 7
if num % 2 == 0:
    print(num, "is Even")
else:
    print(num, "is Odd")
```

## Problem 5: Find the Largest of Three Numbers

```python
a, b, c = 15, 42, 27
largest = max(a, b, c)
print("Largest number:", largest)
```

## Problem 6: Calculate the Factorial of a Number

```python
num = 5
factorial = 1
for i in range(1, num + 1):
    factorial *= i
print(f"Factorial of {num} is {factorial}")
```

## Problem 7: Check if a Number is Prime

```python
num = 11
is_prime = num > 1
for i in range(2, int(num ** 0.5) + 1):
    if num % i == 0:
        is_prime = False
        break
print(f"{num} is {'Prime' if is_prime else 'Not Prime'}")
```

## Problem 8: Reverse a String

```python
text = "Python"
reversed_text = text[::-1]
print("Reversed string:", reversed_text)
```

## Problem 9: Print the Fibonacci Series up to N Terms

```python
n = 8
a, b = 0, 1
fib_series = []
for _ in range(n):
    fib_series.append(a)
    a, b = b, a + b
print("Fibonacci series:", fib_series)
```
