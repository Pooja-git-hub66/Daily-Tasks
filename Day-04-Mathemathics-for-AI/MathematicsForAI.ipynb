```python
#Eigenvalues and Eigenvectots
import numpy as np

# Step 1: Create the matrix
A = np.array([
    [4, 2],
    [1, 3]
])

# Step 2: Find eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Step 3: Print the matrix
print("Matrix:")
print(A)

# Step 4: Print eigenvalues
print("\nEigenvalues:")
print(eigenvalues)

# Step 5: Print eigenvectors
print("\nEigenvectors:")
print(eigenvectors)

# Step 6: Verify the first eigenvector
v1 = eigenvectors[:, 0]
lambda1 = eigenvalues[0]

print("\nA @ v1:")
print(A @ v1)

print("\nlambda1 * v1:")
print(lambda1 * v1)
```

## Probability

```python
import random
roll = random.randint(1,6)
print("You rolled:",roll)
```

```python
import random
rolls = 100000

count_three = 0
count_even = 0
count_greater_than_4 = 0

for i in range(rolls):
    roll = random.randint(1, 6)

    if roll == 3:
        count_three += 1

    if roll % 2 == 0:
        count_even += 1

    if roll > 4:
        count_greater_than_4 += 1

print("Probability of rolling 3:", count_three / rolls)
print("Probability of rolling an even number:", count_even / rolls)
print("Probability of rolling a number greater than 4:", count_greater_than_4 / rolls)
```

```python
#Bayes Theorm

# Prior probability of disease
P_Disease = 10 / 1000

# Probability of testing positive if disease exists
P_Positive_given_Disease = 0.99

# Probability of testing positive
P_Positive = ((10 * 0.99) + (990 * 0.05)) / 1000

# Bayes' Theorem
P_Disease_given_Positive = (
    P_Positive_given_Disease * P_Disease
) / P_Positive

print("Probability of Disease given Positive Test:")
print(P_Disease_given_Positive)
```

```python
# Probabilities
P_spam = 0.20
P_not_spam = 0.80

P_marked_given_spam = 0.95
P_marked_given_not_spam = 0.10

# Probability that an email is marked as spam
P_marked = (
    P_marked_given_spam * P_spam
) + (
    P_marked_given_not_spam * P_not_spam
)

# Bayes' Theorem
P_spam_given_marked = (
    P_marked_given_spam * P_spam
) / P_marked

print("Probability an email is marked as spam:", P_marked)
print("Probability an email is actually spam given it was marked as spam:", P_spam_given_marked)
```
