## 1. Time & Space Complexity / Big-O Notation

- **O(1)** - constant time (e.g. accessing an array element by index)
- **O(log n)** - logarithmic time (e.g. binary search)
- **O(n)** - linear time (e.g. looping through a list once)
- **O(n log n)** - linearithmic time (e.g. merge sort, quick sort)
- **O(n^2)** - quadratic time (e.g. bubble sort, nested loops)

The example below shows an O(n) loop vs an O(n^2) nested loop and times both.

```python
import time

n = 2000
data = list(range(n))

start = time.perf_counter()
total = sum(data)                      # O(n)
print("O(n) sum:", total, "time:", round(time.perf_counter() - start, 5), "s")

start = time.perf_counter()
pairs = 0
for i in data:                          # O(n^2)
    for j in data:
        pairs += 1
print("O(n^2) pair count:", pairs, "time:", round(time.perf_counter() - start, 5), "s")
```

## 3. Linked List (Concept - Singly Linked List)

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def display(self):
        elements = []
        current = self.head
        while current:
            elements.append(current.data)
            current = current.next
        return elements

ll = LinkedList()
for value in [10, 20, 30, 40]:
    ll.append(value)
print("Linked list contents:", ll.display())
```

## 5. Queue (FIFO)

```python
from collections import deque

queue = deque()

queue.append("A")   # enqueue
queue.append("B")
queue.append("C")
print("Queue after enqueues:", list(queue))

front = queue.popleft()  # dequeue
print("Dequeued:", front)
print("Queue after dequeue:", list(queue))
```

## 7. Trees - Binary Search Tree (BST)

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def insert(root, value):
    if root is None:
        return TreeNode(value)
    if value < root.value:
        root.left = insert(root.left, value)
    else:
        root.right = insert(root.right, value)
    return root

def inorder_traversal(root, result=None):
    if result is None:
        result = []
    if root:
        inorder_traversal(root.left, result)
        result.append(root.value)
        inorder_traversal(root.right, result)
    return result

def search_bst(root, value):
    if root is None or root.value == value:
        return root is not None
    if value < root.value:
        return search_bst(root.left, value)
    return search_bst(root.right, value)

root = None
for value in [50, 30, 70, 20, 40, 60, 80]:
    root = insert(root, value)

print("In-order traversal (sorted):", inorder_traversal(root))
print("Search for 40:", search_bst(root, 40))
print("Search for 100:", search_bst(root, 100))
```

## 8. Searching - Linear Search & Binary Search

```python
def linear_search(arr, target):
    for i, val in enumerate(arr):
        if val == target:
            return i
    return -1

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

data = [3, 8, 15, 21, 34, 42, 55, 61, 78]

print("Linear search for 34:", linear_search(data, 34))
print("Binary search for 34:", binary_search(data, 34))
print("Binary search for 99 (not found):", binary_search(data, 99))
```

## 9. Sorting Algorithms

### 9.1 Bubble Sort

```python
def bubble_sort(arr):
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

print(bubble_sort([64, 34, 25, 12, 22, 11, 90]))
```

### 9.2 Selection Sort

```python
def selection_sort(arr):
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

print(selection_sort([64, 25, 12, 22, 11]))
```

### 9.3 Insertion Sort

```python
def insertion_sort(arr):
    arr = arr.copy()
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

print(insertion_sort([12, 11, 13, 5, 6]))
```

### 9.4 Merge Sort

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

print(merge_sort([38, 27, 43, 3, 9, 82, 10]))
```

### 9.5 Quick Sort

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

print(quick_sort([10, 7, 8, 9, 1, 5]))
```

## 10. Recursion

### 10.1 Factorial using Recursion

```python
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)

print("Factorial of 6:", factorial(6))
```

### 10.2 Fibonacci using Recursion

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

fib_series = [fibonacci(i) for i in range(10)]
print("First 10 Fibonacci numbers:", fib_series)
```
