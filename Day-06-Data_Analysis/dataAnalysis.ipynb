```python
#filtering data
import pandas as pd

df = pd.read_csv("students.csv")

print(df)

#Selecting a Single Column
print(df["Name"])

#Selecting Multiple Columns
print(df[["Name", "Marks"]])

#filter rows
print(df[df["Marks"] > 90]) #Students with Marks > 90
print(df[df["Marks"] >= 90]) #Students with Marks >= 90
print(df[df["Age"] < 22]) #Students with Age < 22
print(df[df["City"] == "Delhi"]) #Students from Delhi
print(df[df["City"] != "Delhi"]) #Students not from Delhi
print(df[(df["Age"] > 20) & (df["Marks"] > 85)]) #Multiple Conditions (AND)
print(df[(df["City"] == "Delhi") | (df["Marks"] > 90)]) #Multiple Conditions (OR)
print(df[df["City"].isin(["Delhi", "Mumbai"])]) #Filter Using isin()
print(df.sort_values("Marks")) #sorting data, ascending order
print(df.sort_values("Marks", ascending=False)) #sorting data, descending order
print(df.sort_values(["Age", "Marks"])) #Sort by Multiple Columns
top_students = df.sort_values("Marks", ascending=False)

print(top_students.head(3)) #selecting top 3 students based on marks
```

```python
# groupBy
# create a dataframe
import pandas as pd

data = {
    "Department": ["IT", "HR", "IT", "Sales", "HR", "Sales"],
    "Employee": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank"],
    "Salary": [50000, 40000, 60000, 70000, 45000, 75000]
}

df = pd.DataFrame(data)
print(df.groupby("Department")) #grouping by department
print(df.groupby("Department")["Salary"].mean()) #average salary by department
print(df.groupby("Department")["Salary"].sum()) #sum
print(df.groupby("Department")["Salary"].count()) #count
print(df.groupby("Department")["Salary"].max()) #maximum salary
print(df.groupby("Department")["Salary"].min()) #minimum salary
print(df.groupby("Department")["Salary"].median()) #median salary by department
print(
    df.groupby("Department")["Salary"].agg(
        ["mean", "sum", "max", "min", "count"]
    )
) #multiple aggregations
data = {
    "Department": ["IT", "IT", "HR", "HR", "Sales", "Sales"],
    "Gender": ["F", "M", "M", "F", "M", "F"],
    "Salary": [50000, 60000, 40000, 45000, 70000, 75000]
}

df = pd.DataFrame(data) #group by multiple columns
avg_salary = df.groupby("Department")["Salary"].mean()

print(avg_salary) #reset index

print(df)
```

```python
#Missing Value Handling(NaN)
#Create a DataFrame with Missing Values
import pandas as pd
import numpy as np

data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [20, np.nan, 22, 24],
    "Marks": [90, 85, np.nan, 95]
}

df = pd.DataFrame(data)
print(df.isnull())  #checking for missing values
print(df.isnull().sum()) #count missing values
clean_df = df.dropna()

print(clean_df) #remove rows with missing values
df.dropna(axis=1) #remove columns with missing values
df.fillna(0) #fill missing values with constant
df["Age"] = df["Age"].fillna(df["Age"].mean())

print(df) #fill missing values with mean 
df["Marks"] = df["Marks"].fillna(df["Marks"].median()) #fill missing values with median
df.ffill() #forward fill(ffill)
df.bfill() #backward fill(bfill)
print(df.isnull().sum()) #check if any missing values remain


print(df)
```

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [20, np.nan, 22, 24],
    "Marks": [90, 85, np.nan, 95]
}

df = pd.DataFrame(data)

print("Original Data:")
print(df)

print("\nMissing Values:")
print(df.isnull().sum())

df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Marks"] = df["Marks"].fillna(df["Marks"].median())

print("\nCleaned Data:")
print(df)

print("\nRemaining Missing Values:")
print(df.isnull().sum())
```

```python
# data preprocessing
#Remove Duplicate Rows
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Alice", "David"],
    "Marks": [90, 85, 90, 95]
}

df = pd.DataFrame(data)

print(df)

#rename columns
df.rename(columns={"Marks": "Score"}, inplace=True)

print(df)

#change data type
print(df.dtypes)
df["Score"] = df["Score"].astype(float)
df["Score"] = df["Score"].astype(int)
df["Score"] = df["Score"].astype(str)

#Create New Columns
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Marks": [90, 35, 70]
}

df = pd.DataFrame(data)

#string operations
df["Name"] = df["Name"].str.upper()
df["Name"] = df["Name"].str.lower()
df["Name"] = df["Name"].str.title()
df["Name"] = df["Name"].str.strip()

#basic encoding
df["Gender"] = df["Gender"].map({
    "Male":1,
    "Female":0
})
```

```python
import pandas as pd

data = {
    "Name": ["Alice","Bob","Alice","David"],
    "Marks":[90,35,90,95]
}

df = pd.DataFrame(data)

df = df.drop_duplicates()

df.rename(columns={"Marks":"Score"}, inplace=True)

df["Pass"] = df["Score"] >= 40

df["Name"] = df["Name"].str.upper()

print(df)
```

## feature Engineering basics


```python
#bmi calculation
import pandas as pd

data = {
    "Height":[170,160,180],
    "Weight":[70,60,90]
}

df = pd.DataFrame(data)
df["BMI"] = df["Weight"] / ((df["Height"]/100)**2)

print(df)
```

```python
#bonus
data = {
    "Employee":["A","B","C"],
    "Salary":[40000,50000,60000]
}

df = pd.DataFrame(data)

df["Bonus"] = df["Salary"] * 0.10
# total marks
data = {
    "Math":[80,90,75],
    "Science":[70,95,85]
}

df = pd.DataFrame(data)

df["Total"] = df["Math"] + df["Science"]
#Averge Marks
df["Average"] = (df["Math"] + df["Science"]) / 2
#pass/fail
df["Pass"] = df["Average"] >= 40
print(df)
```
