```python
class Student:
    Department = "AiMl"

    def __init__(self, name, rollNo):
        self.name = name
        self.rollNo = rollNo

Student1=Student("Pooja",107)
print(Student1.name)
print(Student1.Department)
```

```python
class Dog:
    Species = "Canies"

    def __init__(self,name,age):
        self.name=name
        self.age=age

Dog1=Dog("Tommy",6)
print(Dog1.name)
print(Dog1.Species)
print(Dog1.age)
```

```python
class Calc:
    @staticmethod
    def add(a, b):
        return a + b

res = Calc.add(2, 3)
print(res)
```
